```python
import numpy as np
import pandas as pd
```

```python
columns = ['display', 'Test image A', 'Test image B', 'left', 'right', 'work', 'shirk', 'noninspect', 'worked', 'inspected', 'shirked', 'noninspected', 'randomise_trials']
N = 20
ti = ['s0', 's1']
tib = ['i0', 'i1']
wi = ['w0', 'w1']
ni = ['n0', 'n1']
si = ['sh0', 'sh1']
ii = ['in0', 'in1']
rw = ['wr0', 'wr1']
rn = ['nr0', 'nr1']
rs = ['shr0', 'shr1']
ri = ['inr0', 'inr1']
for i in [ti, tib, wi, si, rw, rs, ni, ii, rn, ri]:
    for j in range(2): 
        i[j] = i[j] + '.PNG'
        
w = ['right', 'left']
rwd = [rw, rs]
randomize = 1
```

```python
index = list(np.repeat(0, N/2))
index.extend(list(np.repeat(1, N/2)))
rng = np.random.default_rng()
randorder = rng.choice(list(np.linspace(0, N-1, N, dtype = 'int')), N, 0)
if randomize: 
    iu = []
    for i in randorder: 
        iu.append(index[i])
else: 
    iu = index
```

```python
D, T, L, R, W, S, We, Se, C =['intro', 'attention', 'attention', 'attention', 'attention', 'attention'], ['', '', '', '', '', ''],  ['', '', '', '', '', ''], ['', '', '', '', '', ''],  ['', '', '', '', '', ''],  ['', '', '', '', '', ''],  ['', '', '', '', '', ''],  ['', '', '', '', '', ''], ['', '', '', '', '', '']
p = [0.5]
r = [0, 1, 1, 1, 1, 1]
for i in range(N): 
    D.append('trials')
    T.append(ti[iu[i]])
    p.append(0.5)
    if iu[i] == 0: 
        L.append(0)
        R.append(1)
    else:
        L.append(1)
        R.append(0)
    W.append(1)
    S.append(0)
    We.append(wi[iu[i]])
    Se.append(si[iu[i]])
    C.append(0)
    r.append(1)
TB, Nn, Ne,Ie = ['', '', '', '', '', ''], ['', '', '', '', '', ''],  ['', '', '', '', '', ''],  ['', '', '', '', '', '']
for i in range(N): 
    TB.append(tib[iu[i]])
    Nn.append(0)
    Ne.append(ni[iu[i]])
    Ie.append(ii[iu[i]])
data = pd.DataFrame(np.transpose([D, T, TB, L, R, W, S, Nn, We, Ie, Se, Ne, r]), columns = columns)
emp = ['', '', '', '', '', '']

emp.extend(list(np.repeat('1, 2, 3, 4, 5, 6, 7', N)))
data['Likert'] = emp
tracer = list(np.repeat(1, N+5))
tracer.append(0)
data['trace'] = tracer
data.to_csv('../codes/sprdst1M.csv')
```
