```python
import pandas as pd
```

```python
d = pd.read_csv('drive.csv')
d
```

```python
names = d['File name']
links = d['Download Link']
```

```python
dictionary = {}
for n in range(len(names)): 
    modstr = links[n]
    modstr = '\"' + modstr[0:-8] + 'view\"'
    modname = 'var ' + names[n][0:-4]
    dictionary[modname] = modstr
```
